### Sınıflandırma Algoritmalarının Uygulanması

```python
x = veriler.iloc[:,0:13].values #bağımsız değişken
y = veriler.iloc[:,13:].values #bağımlı değişken
```

```python
#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler   #veri ölçeklendirme

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski')
knn.fit(X_train,y_train.ravel())

y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
confidence = knn.score(X_test, y_test)
print(confidence)
```

```python
from sklearn.model_selection import GridSearchCV
grid_params = {
    "n_neighbors" : [3,5,11,19],
    "weights" : ["uniform", "distance"],
    "metric" :["euclidean", "manhattan"]
}

gs = GridSearchCV(KNeighborsClassifier(),
                 grid_params,
                 verbose= 1,
                 cv = 3)

gs_results = gs.fit(X_train, y_train.ravel())
```

```python
gs_results.best_score_
```

```python
gs_results.best_estimator_
```

```python
gs_results.best_params_
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train.ravel())
```

```python
y_pred = svc.predict(X_test)
```

```python

cm = confusion_matrix(y_test,y_pred)
print('SVC')
print(cm)
```

```python
confidence = svc.score(X_test, y_test)
print(confidence)
```
